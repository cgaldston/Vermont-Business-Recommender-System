# Begin training Models

We can start by building tuples like the ones in assignment 1 with just the user, movie, and rating. We can then add back in other features using the metadata from the movies_data dataset. 

```python
def MSE(predictions, labels):
    differences = [(x-y)**2 for x,y in zip(predictions,labels)]
    return sum(differences) / len(differences)
```

```python
trainset, testset = train_test_split(ratings_movies, test_size=.25, random_state=48)
```

```python
dataTrain = list(trainset[['user_id', 'movie_id', 'rating_val']].itertuples(index=False, name=None))
dataTest = list(testset[['user_id', 'movie_id', 'rating_val']].itertuples(index=False, name=None))
```

### Baseline

```python
trainRatings = [r[2] for r in dataTrain]
globalAverage = sum(trainRatings) * 1.0 / len(trainRatings)
```

```python
globalAverage
```

```python
validMSE = 0
for u,m,r in dataTest:
    sse = (r - globalAverage)**2
    validMSE += sse

validMSE /= len(dataTest)

print("Baseline MSE on test set (predictions using training average) = " + str(validMSE))
```
