# Models

```python
# Remove null ratings
data = [record for record in data if record.get('rating') is not None]
```

```python
def MSE(predictions, labels):
    differences = [(x-y)**2 for x,y in zip(predictions,labels)]
    return sum(differences) / len(differences)
```

```python
trainset, testset = train_test_split(data, test_size=.25, random_state=48)
```

```python
trainRatings = [r['rating'] for r in trainset]
trainingAverage = sum(trainRatings) * 1.0 / len(trainRatings)
```

```python
validMSE = 0
for d in testset:
    sse = (d['rating'] - trainingAverage)**2
    validMSE += sse

validMSE /= len(testset)

print("Baseline MSE on test set (predictions using training average) = " + str(validMSE))
```

### Temporal

```python
ratingsTime = []
for d in data:
    ratingsTime.append((d["time"], d["rating"]))
```

```python
ratingsTime.sort()
```

```python
wSize = 5000
x = [r[0] for r in ratingsTime] # as raw times
y = [r[1] for r in ratingsTime] # ratings
x_datetime = [datetime.fromtimestamp(t / 1000) for t in x]
xu = [time.mktime(d.timetuple()) for d in x_datetime]
```

```python
xSum = sum(xu[:wSize])
ySum = sum(y[:wSize])
sliding = []
```

```python
for i in range(wSize,len(x)-1):
    xSum += xu[i] - xu[i-wSize]
    ySum += y[i] - y[i-wSize]
    sliding.append((xSum*1.0/wSize,ySum*1.0/wSize))
```

```python
X = [a[0] for a in sliding]
Y = [a[1] for a in sliding]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(X[::2500],Y[::2500], label="5000 rating window", color='k')
plt.xticks([X[600], X[-350]], [x_datetime[wSize+600].strftime("%d/%m/%Y"), x_datetime[-350].strftime("%d/%m/%Y")])
plt.xlim(X[0], X[-1])
plt.ylabel("Rating")
plt.xlabel("Date")
plt.legend(loc="best",fontsize=8)
plt.title("Ratings over time (sliding windows)")
plt.show()
```

```python
from datetime import datetime, timedelta

# Define the start and end dates for the x-axis
start_date = datetime(2016, 1, 1)
end_date = datetime.fromtimestamp(X[-1])  # Convert last x value to datetime

# Generate more x-axis tick locations (e.g., every 6 months)
num_ticks = 10  # Number of ticks you want on the x-axis
tick_dates = [start_date + (end_date - start_date) * i / (num_ticks - 1) for i in range(num_ticks)]
tick_locations = [d.timestamp() for d in tick_dates]  # Convert to UNIX timestamps
tick_labels = [d.strftime("%b %Y") for d in tick_dates]  # Format dates as "Month Year"

# Stretch the plot and customize the x-axis
plt.figure(figsize=(12, 6))  # Increase the figure size
plt.plot(X[::1000], Y[::1000], label="5000 rating window", color='k')

# Set x-axis limits and ticks
plt.xlim(start_date.timestamp(), X[-1])  # Focus on data from 2016 onwards
plt.xticks(tick_locations, tick_labels, rotation=45)  # Add tick labels with rotation

plt.ylabel("Rating")
plt.xlabel("Date")
plt.legend(loc="best", fontsize=8)
plt.title("Ratings over time (sliding windows)")
plt.tight_layout()  # Ensure everything fits nicely
plt.show()
```

```python
userIDs = {}
itemIDs = {}
interactions = []
interactionsPerUser = defaultdict(list)

for d in data:
    u = d['user_id']
    i = d['gmap_id']
    t = d['time']
    r = d['rating']
    t = int(t) // 1000
    if not u in userIDs: userIDs[u] = len(userIDs)
    if not i in itemIDs: itemIDs[i] = len(itemIDs)
    interactions.append((t,u,i,r))
    interactionsPerUser[u].append((t,i,r))
```

```python
nUsers,nItems = len(userIDs),len(itemIDs)
```

```python
nUsers,nItems
```

```python
time_data = np.array([data[i]['time'] for i in range(len(data))]).reshape(-1, 1)
scaler = MinMaxScaler()
time_data_normalized = scaler.fit_transform(time_data)
```

```python
X = sp.lil_matrix((len(data), nUsers + nItems + 1))
```

```python
for i in range(len(data)):
    user = userIDs[data[i]['user_id']]
    item = itemIDs[data[i]['gmap_id']]
    time_normalized = time_data_normalized[i, 0]
    
    X[i,user] = 1 # One-hot encoding of user
    X[i,nUsers + item] = 1 # One-hot encoding of item
    X[i, nUsers + nItems] = time_normalized
```

```python
y = np.array([d['rating'] for d in data])
```

```python
fm = als.FMRegression(n_iter=100, init_stdev=0.1, rank=5, l2_reg_w=0.1, l2_reg_V=0.5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=48 
)
```

```python
fm.fit(X_train, y_train)
```

```python
y_pred = fm.predict(X_test)
```

```python
mse_fm = MSE(y_pred, y_test)
print("FM model = " + str(mse_fm))
```
